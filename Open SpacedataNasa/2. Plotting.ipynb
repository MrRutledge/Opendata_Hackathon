Let's plot some of the UFO sightings data

```python
df = pd.read_csv(os.path.join(data_directory,"UFO_Sightings_Global.csv"))
```

```python
df.head()
```

```python
df.head()
```

### Categorical scatterplots

```python
g = sns.catplot(x='Shape', 
                y='duration (s)', 
                data=df, height=8, aspect=1.5)
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-90)
```

```python
g = sns.catplot(x='Shape', 
                y='duration (s)', 
                data=df[1:50],
                jitter=False, height=8, aspect=1.5) 
g.set_xticklabels(rotation=-90)
```

### Boxplots

```python
g = sns.catplot(x='Shape', 
                y='duration (s)', 
                data=df[1:50],
                kind="box",
                height=8, 
                aspect=1.5)

g.set_xticklabels(rotation=-45)
```

### Violin Plots

```python
g = sns.catplot(x='Shape', 
                y='duration (s)', 
                data=df.sample(6),
                kind="violin",
                height=8, 
                aspect=1.5) 
g.set_xticklabels(rotation=-45)
```

Let's take another dataset from the list.

```python
df2 = pd.read_csv(os.path.join(data_directory, "SolarSystemAndEarthquakes.csv"))
```

```python
df2.head()
```

### Scatterplot

```python
sns.lmplot('earthquake.longitude', 'earthquake.latitude', data=df2, fit_reg=False,height=7, aspect=1.5)
```

### Joint Distribution Plot

```python
sns.jointplot(x='earthquake.longitude',
              y='earthquake.mag',
              data=df2
              ,height=10) 
             # ,ratio=2)
```

We can also plot a linear regression with the marginal distributions and apply some style to the plots.

```python
sns.set(style="darkgrid")
sns.jointplot(x='earthquake.longitude',
              y='earthquake.mag',
              kind='reg',
              color="g",
              data=df2, height=10)
```

### Barplot

```python
sns.barplot(data = df2.head(),y= 'earthquake.place',x = 'earthquake.mag')
plt.ylabel('Earthquake Magnitude ', fontsize=12)
plt.xlabel('Place ', fontsize=12)
#plt.xticks(rotation='vertical')
```

```python
#df2[df2['earthquake.mag']>7]
```

```python
sns.barplot(data = df2[df2['earthquake.mag']>7].sample(20),y = 'earthquake.place',x = 'earthquake.mag')
plt.xlabel('Earthquake Magnitude ', fontsize=12)
plt.ylabel('Place ', fontsize=12)
```

### Density Plot

```python
sns.kdeplot(df2['earthquake.mag'])
```

```python
sns.kdeplot(df2['earthquake.longitude'], df2['earthquake.latitude'])
```

### Heatmap plot
